```python
import matplotlib.pyplot as plt                     #include libs
from scipy.optimize import curve_fit
from IPython.display import display, Math, Latex
import numpy as np

plt.rcParams["font.family"] = "monospace"

def mapping(x, k, a):                #Function of approximating give to curve_fit
    return a + k*x

plt.figure(figsize=(11,11))                                #Create graphic in matplotlib
plt.title(r"Зависимость разности длин от диаметра крайнего отверстия")
plt.ylabel(r"Разность длин $\delta l$, мм")
plt.xlabel(r"Диаметр отверстия $d$, мм")

#-------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------

#processing data

x = []
y1 = []
y2 = []

#FILE "data.txt" !!!!!!!!!!!! YOU NEED TO UPLOAD IT TO COLAB

file = open('data.txt', 'r')                 #open file with data
while (line := file.readline()):              # every line have x, y, y1 ...
  s = line.split()
  if (len(s) != 0):
    x.append(float(s[1].replace(',','.')))    #collect in x[] y1[] y2[] with change "," -> "." (if excel)
    y1.append(float(s[0].replace(',','.'))*1000)

k1 = 0                                        #create coeffs all in function

coeffs,_ = curve_fit(mapping, x, y1)          #give func and our measurements
k1 = coeffs[0]                                   #it returns array of aproximating coeffs
a = coeffs[1]
y1_fit = []
for i in range(len(x)):
  y1_fit.append(a + k1 * x[i])                    #with coeffs make array of Approximating data
#                 ^
#                 |
#                 function need to write
#                 """"""""""""""""""""""

plt.plot(x, y1, 'r^', label='')    #triangles with measurments
plt.plot(x, y1_fit, color = 'y', label = f"Прямая с аппроксимацией) " r'$\delta l = a + k \cdot d$' f", k = {k1:.3f}, a = {a:.3f}") #approximating praphic

#----------------------------------------------------------------------------------------
#----------------------------------------------------------------------------------------

plt.grid(visible = True, which='major', axis='both', alpha=1)           #end of matplotlib
plt.grid(visible = True, which='minor', axis='both', alpha=1)           #show graphic
plt.legend()
plt.show()
plt.savefig('graphic.png')
file.close()
```
